```python
# Corrigo los tipos de datos , convertir cadenas a fechas.
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
# Crear nuevas columnas extrayendo información de la columna 'mes_venta'
df['Month'] = df['Date'].dt.month

print("Nueva columna creada:")
df['Month'].head(5)
```

```python
# Cambiar el numero por mes del año
month_map = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}
df['Month'] = df['Month'].map(month_map)

print("Columna 'Month' convertida a nombres de mes:")
print(df['Month'].head())
```

```python
# Clasifica los datos en categorías relevantes.
# Encontrando el índice del producto con mayor venta por categoría
idx = df.groupby('Product Category')['Total Amount'].idxmax()

# 'groupby' agrupa el DataFrame por la columna 'Product Category'
# 'idxmax' devuelve el índice de la fila con el valor máximo de ventas dentro de cada grupo (categoría)
# Esto nos da los índices del producto con mayor venta en cada categoría
# Luego, usamos 'loc' para seleccionar las filas correspondientes a esos índices
top_productos = df.loc[idx, ['Product Category', 'Gender', 'Total Amount']]

# Imprimiendo los resultados
print("Genero que mas compra una Categoría:")
print(top_productos)
```

### Clasifica los datos: Crea una columna que clasifique las ventas en categorías significativas (e.g., ‘Alta’, ‘Media’, ‘Baja’).

```python
# Creamos la nueva columna dividiendo en 3 grupos iguales (terciles)
df['Sales Category'] = pd.qcut(df['Total Amount'], q=3, labels=['Low', 'Med', 'Hig'])
```

### Función personalizada: Aplica funciones personalizadas para realizar análisis específicos que no se pueden lograr con las funciones de agregación estándar.

Uso de Apply para calcular impuesto especifico por categoria

```python
def cal_tax (row):
    if (row['Product Category'] == "Beauty"):
        return row['Total Amount'] * 0.20
    elif (row['Product Category'] == "Clothing"):
        return row['Total Amount'] * 0.15
    elif (row['Product Category'] == "Electronics"):
        return row['Total Amount'] * 0.10
    else:
        return 0.0

# Aplicar la función a cada fila del DataFrame
df['Taxes'] = df.apply(cal_tax, axis=1)

df
```

```python
# Calcular el promedio de 'Total Amount' para cada 'Sales Category'
# La Serie 'mean_by_category' será alineada con el index del DataFrame
mean_by_category = df.groupby('Sales Category')['Total Amount'].transform('mean')

def Standard_desviation_by_category (row):
    # 'row.name' obtine el index original de la fila actual, para recibir el promedio
    category_mean = mean_by_category.loc[row.name]

    # Calcular la desviacion de la fila

    return row['Total Amount'] - category_mean

# Aplicar la función a cada fila del DataFrame
df['Standard Desviation'] = df.apply(Standard_desviation_by_category, axis=1)


df
```
